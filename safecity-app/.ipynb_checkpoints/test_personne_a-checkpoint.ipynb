```python
import sys
sys.path.insert(0, '.')

import pandas as pd
from pipeline.fetchers.crimes_fetcher import CrimesFetcher
from pipeline.fetchers.insee_fetcher import INSEEFetcher
from pipeline.fetchers.geo_fetcher import GeoFetcher
from pipeline.storage import save_crimes_data, save_insee_data, save_merged_data

print("‚úÖ Tous les imports OK !")
```

```python
print("="*70)
print("TEST PERSONNE A - √âTAPE 1 : CRIMES FETCHER")
print("="*70 + "\n")

crimes_fetcher = CrimesFetcher()

# Charger depuis le fichier
crimes_df = crimes_fetcher.fetch_from_file()

# Nettoyer
crimes_cleaned = crimes_fetcher.clean_data()

print(f"\nüìä Resume crimes :")
print(crimes_fetcher.get_summary())

print(f"\nüìã Apercu donnees crimes :")
print(crimes_cleaned.head(20))

print(f"\nüìä Info crimes :")
print(crimes_cleaned.info())

print(f"\nüìà Statistiques :")
print(crimes_cleaned.describe())
```

```python
print("="*70)
print("TEST PERSONNE A - √âTAPE 2 : INSEE FETCHER")
print("="*70 + "\n")

insee_fetcher = INSEEFetcher()

# Charger depuis le fichier
insee_df = insee_fetcher.fetch_from_file()

# Nettoyer
insee_cleaned = insee_fetcher.clean_data()

print(f"\nüìä Resume INSEE :")
print(insee_fetcher.get_summary())

print(f"\nüìã Apercu donnees INSEE :")
print(insee_cleaned.head(20))

print(f"\nüìä Info INSEE :")
print(insee_cleaned.info())

print(f"\nüìà Statistiques :")
print(insee_cleaned.describe())
```

```python
print("="*70)
print("TEST PERSONNE A - √âTAPE 3 : GEO FETCHER")
print("="*70 + "\n")

geo_fetcher = GeoFetcher()

# Charger depuis le fichier
geo_data = geo_fetcher.fetch_from_file()

print(f"\nüó∫Ô∏è Resume GeoJSON :")
print(geo_fetcher.get_summary())

if geo_data is not None:
    print(f"\nüìã Apercu GeoJSON :")
    features = geo_data.get("features", [])
    print(f"   Total features : {len(features)}")
    
    if features:
        print(f"\n   Premi√®re feature (d√©partement) :")
        first = features[0]
        print(f"   Properties : {first.get('properties')}")
        print(f"   Geometry type : {first.get('geometry', {}).get('type')}")
        
        print(f"\n   Tous les codes d√©partement :")
        codes = [f.get("properties", {}).get("DDEP_C_COD") for f in features]
        print(f"   {sorted(codes)[:20]}...")
```

```python
print("="*70)
print("TEST PERSONNE A - √âTAPE 4 : FUSION + STOCKAGE")
print("="*70 + "\n")

# Pr√©parer les donn√©es pour fusion
print("Pr√©paration des donn√©es...")

# Renommer colonnes crimes si n√©cessaire
if "Code_departement" in crimes_cleaned.columns:
    crimes_cleaned = crimes_cleaned.rename(columns={"Code_departement": "departement"})

# Fusionner crimes + INSEE
print("Fusion crimes + INSEE...")
merged_df = crimes_cleaned.merge(
    insee_cleaned[["departement", "annee", "population", "nom_departement"]],
    on=["departement", "annee"],
    how="inner"
)

print(f"‚úÖ Donnees fusionnees : {len(merged_df)} lignes")

# Sauvegarder en Parquet
print("\nSauvegarde en Parquet...")
save_crimes_data(crimes_cleaned)
save_insee_data(insee_cleaned)
save_merged_data(merged_df)

print(f"\nüìä Apercu donnees fusionnees :")
print(merged_df.head(20))

print(f"\nüìà Statistiques finales :")
print(f"  - Total lignes : {len(merged_df)}")
print(f"  - Colonnes : {list(merged_df.columns)}")
print(f"  - Ann√©es : {sorted(merged_df['annee'].unique().tolist())}")
print(f"  - D√©partements : {merged_df['departement'].nunique()}")
print(f"  - Types crimes : {merged_df['indicateur'].nunique()}")

print("\n" + "="*70)
print("‚úÖ PERSONNE A TERMIN√âE - DONN√âES PR√äTES !")
print("="*70)
```
