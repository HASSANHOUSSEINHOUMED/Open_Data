```python
import sys
sys.path.insert(0, '.')

import pandas as pd
from utils.data import load_merged_data, load_geojson
from utils.maps import SafeCityMapper
from utils.charts import SafeCityCharts

print("‚úÖ Tous les imports OK !")

# Charger les donnees
print("\nüì• Chargement des donnees...")
merged_df = load_merged_data()
geojson_data = load_geojson()

print(f"‚úÖ Donnees chargees : {len(merged_df)} lignes")
print(f"‚úÖ GeoJSON charge : {len(geojson_data['features'])} features")
```

```python
print("="*70)
print("TEST PERSONNE B - √âTAPE 1 : CARTOGRAPHIE FOLIUM")
print("="*70 + "\n")

# Initialiser le mapper
mapper = SafeCityMapper(merged_df, geojson_data)

print(f"üìä Resume mapper :")
print(mapper.get_summary())

# Test 1 : Creer une carte choropl√®the pour 2024 (tous crimes)
print("\n\n--- TEST 1 : Carte choropl√®the 2024 (tous crimes) ---")
map_2024 = mapper.create_choropleth_map(year=2024)
map_2024.save("map_2024_all_crimes.html")
print(f"‚úÖ Carte sauvegardee : map_2024_all_crimes.html")

# Test 2 : Creer une carte pour homicides 2024
print("\n--- TEST 2 : Carte choropl√®the 2024 (homicides) ---")
map_homicides = mapper.create_choropleth_map(year=2024, crime_type="Homicides")
map_homicides.save("map_2024_homicides.html")
print(f"‚úÖ Carte sauvegardee : map_2024_homicides.html")

print("\n" + "="*70)
print("‚úÖ CARTOGRAPHIE OK - Fichiers HTML g√©n√©r√©s")
print("="*70)
```

```python
print("="*70)
print("TEST PERSONNE B - √âTAPE 2 : GRAPHIQUES PLOTLY")
print("="*70 + "\n")

# Initialiser le charteur
charts = SafeCityCharts(merged_df)

print(f"üìä Resume charts :")
summary = charts.get_summary()
print(f"  - Ann√©es : {summary['years']}")
print(f"  - Types crimes : {len(summary['crime_types'])} types")
print(f"  - D√©partements : {len(summary['departments'])} depts")

# Test 1 : Graphique evolution temporelle (France)
print("\n\n--- TEST 1 : Evolution temporelle (tous crimes) ---")
fig_timeline = charts.create_timeline_chart()
fig_timeline.show()

# Test 2 : Top 10 departements 2024
print("\n--- TEST 2 : Top 10 departements 2024 ---")
fig_top10 = charts.create_bar_chart_top_departments(year=2024, top_n=10)
fig_top10.show()

# Test 3 : Comparaison Paris vs Lyon
print("\n--- TEST 3 : Comparaison Paris vs Rhone (Lyon) ---")
fig_comparison = charts.create_comparison_chart(year=2024, dept1="Paris", dept2="Rhone")
fig_comparison.show()

# Test 4 : Pie chart distribution crimes
print("\n--- TEST 4 : Distribution des crimes par type (2024) ---")
fig_pie = charts.create_pie_chart(year=2024)
fig_pie.show()

print("\n" + "="*70)
print("‚úÖ GRAPHIQUES OK - 4 visualisations g√©n√©r√©es")
print("="*70)
```
