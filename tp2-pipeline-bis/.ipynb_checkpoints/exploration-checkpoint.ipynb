```python
# Extraire les donnees interessantes du DataFrame

import pandas as pd
import json

parquet_file = "C:\\tp2-pipeline-bis\\data\\processed\\weather_20251216_170346.parquet"
df = pd.read_parquet(parquet_file)

print("="*80)
print("DONNEES METEO ENRICHIES - VUE SIMPLIFIEE")
print("="*80 + "\n")

# Extraire les infos principales du JSON imbrique
df_simplified = pd.DataFrame({
    'Ville': df['city'],
    'Adresse': df['address_label'],
    'Code Postal': df['postal_code'],
    'Latitude': df['latitude'],
    'Longitude': df['longitude'],
    'Score Geocodage': df['geocoding_score'].round(3),
})

# Extraire la temperature depuis la colonne 'current'
df_simplified['Temperature (Â°C)'] = df['current'].apply(
    lambda x: x.get('temperature_2m') if isinstance(x, dict) else None
)
df_simplified['Humidite (%)'] = df['current'].apply(
    lambda x: x.get('relative_humidity_2m') if isinstance(x, dict) else None
)
df_simplified['Vitesse Vent (km/h)'] = df['current'].apply(
    lambda x: x.get('wind_speed_10m') if isinstance(x, dict) else None
)

print(df_simplified.to_string(index=False))

print("\n" + "="*80)
print("RESUME")
print("="*80)
print(f"Total de villes analysees : {len(df)}")
print(f"Score geocodage moyen : {df['geocoding_score'].mean():.3f}")
print(f"Temperature moyenne : {df_simplified['Temperature (Â°C)'].mean():.1f}Â°C")
```
